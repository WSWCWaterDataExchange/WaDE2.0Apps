```python
WSWCBenUseColorDict = {
"Agricultural " : "Yellow",
"Commercial " : "DarkGreen",
"Domestic " : "Blue",
"Fire " : "Crimson",
"Industrial " : "BlueViolet",
"Mining " : "Gold",
"Municipal " : "Black",
"Power " : "DarkOrange",
"Recreation " : "Fuchsia",
"Snow Making " : "MintCream",
"StockWatering " : "DarkTurquoise",
"Wildlife " : "PaleGreen",
"Other " : "SlateGray"}


def assignWSWCBUColor(colrowValue):
    if colrowValue == '' or pd.isnull(colrowValue):
        return "SlateGray"
    else:
        String1 = colrowValue  # remove whitespace chars
        try:
            outList = WSWCBenUseColorDict[String1]
        except:
            outList = "SlateGray"
    return outList

df['WBenUseColor'] = df.apply(lambda row: assignWSWCBUColor(row['WSWCBeneficialUse']), axis=1)
```

```python
#Removing NaN, and missing (999) values from AllocaitonAmount, AllocaitonMaxium, IrrigatedAcerage.
df['AllocationAmount'] = df['AllocationAmount'].fillna(0)
df['AllocationAmount'] = df['AllocationAmount'].replace(999, 0)

df['AllocationMaximum'] = df['AllocationMaximum'].fillna(0)
df['AllocationMaximum'] = df['AllocationMaximum'].replace(999, 0)

df['IrrigatedAcreage'] = df['IrrigatedAcreage'].fillna(0)
df['IrrigatedAcreage'] = df['IrrigatedAcreage'].replace(999, 0)
```

```python
#Outbound DataFrame
#outdf = pd.DataFrame(index = df.index)
outdf = pd.DataFrame()

outdf['IndexNum'] = df.index
outdf['SiteUUID'] = df['SiteUUID'].astype(str) 
outdf['Lat'] = df['Latitude'].astype(float) 
outdf['Long'] = df['Longitude'].astype(float) 
outdf['AA_ID'] = df['AllocationAmountID'].astype(str)
outdf['SN_ID'] = df['AllocationNativeID'].astype(str) 
outdf['State'] = df['State'].astype(str) 
outdf['AA_CFS'] = df['AllocationAmount'].astype(float) 
outdf['AA_AF'] = df['IrrigatedAcreage'].astype(float) 
outdf['AAM_AF'] = df['AllocationMaximum'].astype(float) 
outdf['PD'] = df['Date']
outdf['WSN'] = df['WaterSourceName'].astype(str) 
outdf['WBenUse'] = df['WSWCBeneficialUse'].astype(str)
outdf['WBenUseColor'] = df['WBenUseColor'].astype(str)

outdf.reset_index()
```
