```python
#Changing datatype of used date fields. 
df['Date'] = pd.to_datetime(df['Date'], errors = 'coerce')
df['Date'] = pd.to_datetime(df["Date"].dt.strftime('%m/%d/%Y'))
```

```python
#Removing NaN, and missing (999) values from AllocaitonAmount, AllocaitonMaxium, IrrigatedAcerage.
df['AllocationAmount'] = df['AllocationAmount'].fillna(0)
df['AllocationAmount'] = df['AllocationAmount'].replace(999, 0)

df['AllocationMaximum'] = df['AllocationMaximum'].fillna(0)
df['AllocationMaximum'] = df['AllocationMaximum'].replace(999, 0)

df['IrrigatedAcreage'] = df['IrrigatedAcreage'].fillna(0)
df['IrrigatedAcreage'] = df['IrrigatedAcreage'].replace(999, 0)
```

```python
WSWCBenUseDict = {
"Irrigation" : "Agricultural",
"Agriculture" : "Agricultural",
"Commercial" : "Commercial",
"Domestic" : "Domestic",
"72-12-1 domestic one household" : "Domestic",
"72-12-1 domestic and livestock watering" : "Domestic",
"72-12-1 multiple domestic households" : "Domestic",
"Subdivision" : "Domestic",
"Mobile home parks" : "Domestic",
"Fire" : "Fire",
"Fire Protection" : "Fire",
"Industrial" : "Industrial",
"Oil production" : "Industrial",
"Highway construction" : "Industrial",
"Cemetery" : "Industrial",
"Poultry and egg operation" : "Industrial",
"Dairy operation" : "Industrial",
"Construction" : "Industrial",
"Domestic construction" : "Industrial",
"Mining" : "Mining",
"Municipal" : "Municipal",
"Power" : "Power",
"Power Generation" : "Power",
"Recreation" : "Recreation",
"Snow Making" : "Snow Making",
"Stockwatering" : "StockWatering",
"Stock" : "StockWatering",
"Wildlife" : "Wildlife",
"Fish and game propogation" : "Wildlife",
"Fish And Wildlife" : "Wildlife"}


def assignWSWCBU(colrowValue):
    if colrowValue == '' or pd.isnull(colrowValue):
        outList = "Other"
    else:
        String1 = colrowValue  # remove whitespace chars
        try:
            outList = WSWCBenUseDict[String1]
        except:
            outList = "Other"
    return outList

df['WSWCBeneficialUse'] = df.apply(lambda row: assignWSWCBU(row['BeneficialUseCV']), axis=1)
```

```python
#Creating API call for SiteUUID
#Need to include qutation marks into string, have to do a little string maniuplation with ' and \" for it to work.

def assignAPICallSiteUUID(colrowValue):
    String1 = str(colrowValue)
    outstring = '<a href=\"https://wade-api-qa.azure-api.net/v1/SiteAllocationAmounts?SiteUUID= '+ String1 + '\">Link</a>'
    return outstring 

df['APICall'] = df.apply(lambda row: assignAPICallSiteUUID(row['SiteUUID']), axis=1)
```

```python
#Outbound DataFrame
outdf = pd.DataFrame()

# outdf['StateDefinedBeneficialUse'] = df['BeneficialUseCV'].astype(str) 
# outdf['WaterSourceName'] = df['WaterSourceName'].astype(str) 
# outdf['DataSourceOrganizationID'] = df['OrganizationUUID'].astype(str)

outdf['AA_ID'] = df['AllocationAmountID'].astype(float) 
outdf['SiteUUID'] = df['SiteUUID'].astype(str)  #new
outdf['State'] = df['State'].astype(str)
outdf['SN_ID'] = df['AllocationNativeID'].astype(str)
outdf['AA_CFS'] = df['AllocationAmount'].astype(float) 
outdf['AA_AF'] = df['IrrigatedAcreage'].astype(float) 
outdf['AAM_AF'] = df['AllocationMaximum'].astype(float) 
outdf['PD'] = df['Date']
outdf['WBenUse'] = df['WSWCBeneficialUse'].astype(str) 
outdf['Lat'] = df['Latitude'].astype(float) 
outdf['Long'] = df['Longitude'].astype(float) 
outdf['APILink'] = df['APICall'].astype(str)

outdf.reset_index()
```

```python
#Remove bad inputs
#some of the Utah sites have goofy long and lat values.

mask = outdf.loc[(outdf["Lat"].isnull()) | 
                 (outdf["Lat"] < 10) | 
                 (outdf["Long"] > -100)]

if len(mask.index) > 0:
    dropIndex = outdf.loc[(outdf["Lat"].isnull()) | 
                          (outdf["Lat"] < 10) | 
                          (outdf["Long"] > -100) ].index
    outdf = outdf.drop(dropIndex)
    outdf = outdf.reset_index(drop=True)
```

```python
#The Ouput
#The Shiny app is responding best to UTF-8 encoding.

mask.to_csv('RemovedInputs.csv', index=False)  # The rows that were removed from output.
outdf.to_csv('AllInput_verA.csv', index=False, encoding="UTF-8")  # The output itself.
```
